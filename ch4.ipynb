```python
import pandas as pd
```

```python
df = pd.read_csv('data/stock_data.csv')
df.columns = ['date', 'AAPL', 'MSFT', 'XOM', 'SPX']
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date', ascending=True)
df
```

```python
df.set_index('date', inplace=True)
```

```python
daily_returns = df.pct_change().dropna()
daily_returns
```

```python
daily_returns.reset_index(inplace=True)
```

```python
daily_returns['year'] = pd.DatetimeIndex(daily_returns['date']).year
daily_returns
```

```python
yearly_returns = daily_returns.groupby('year').sum()
yearly_returns
```
